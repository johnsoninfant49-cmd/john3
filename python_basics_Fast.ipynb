```python
def calc(num1, num2, ops):
  if ops =='+':
    print(f'sum of num1 and num2 is {num1+num2}')
  elif ops =='-':
    print(f'diff of num1 and num2 is {num1-num2}')
  elif ops =='*':
    print(f'muilt of num1 and num2 is {num1*num2}')
  elif ops =='/':
    print(f'div of num1 and num2 is {num1/num2}')
```

```python
calc(56,76,'*')
```
